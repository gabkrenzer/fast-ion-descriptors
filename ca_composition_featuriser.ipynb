```python
from elementembeddings.core import Embedding
from elementembeddings.composition import composition_featuriser
import pandas as pd
```

```python
# Load the magpie data
magpie = Embedding.load_data('magpie')
```

```python
# Print out some of the properties of the ElementEmbeddings class
print(f'The magpie representation has embeddings of dimension {magpie.dim}') 
print(f'The magpie representation contains these elements: \n {magpie.element_list}') # prints out all the elements considered for this representation
print(f'The magpie representation contains these features: \n {magpie.feature_labels}') # Prints out the feature labels of the chosen representation
```

```python
data = pd.read_csv('data.csv')
data
```

```python
df_formula = data['formula_pretty_reduced']

df_formula
```

```python
df_featurised = composition_featuriser(df_formula, embedding="magpie", stats=["mean","sum"])

df_featurised
```

```python
df_mat_id = data['material_id']

df_featurised = pd.concat([df_mat_id, df_featurised], axis = 1)

df_featurised
```

```python
missing_id = ['mp-867700', 'mp-613442', 'mp-867699', 'mp-867677', 'mp-561623', 'mp-867712']

for mat in missing_id:
    df_featurised.drop(df_featurised[df_featurised.material_id == mat].index, inplace=True)
    df_featurised.reset_index(inplace=True, drop=True)
    
df_featurised
```

```python
df_featurised.to_csv('composition_feature_mac.csv')
```
