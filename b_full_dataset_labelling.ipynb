```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
diff = pd.read_csv('data_kasper_full.csv') #Kasper's data obtained from m3gnet
diff.rename(columns={'Unnamed: 0': 'material_id'}, inplace=True)
```

```python
diff
```

```python
data = pd.read_csv('data_mac.csv')

for col_name in ['0.05', '0.1', '0.3', '0.4', '0.6', '1e-05', '0.0001', '0.01']:
    data.drop(col_name, axis=1, inplace=True)
    
data.rename(columns={'Unnamed: 0': 'index'}, inplace=True)
```

```python
data
```

```python
def match(temp):
    temp_s = str(temp)
    #diff[temp_s] = diff[temp_s].fillna(-1e-5)
    match = pd.merge(diff[['material_id', temp_s]], 
                     data[['index', 'material_id', 'formula_pretty_reduced', 'spacegroup_symbol', 
                           'fav', 'fav_tot', '298', '298_tot', '0.2', '0.5', '0.001']], 
                     on=['material_id'], how='inner')
    return match
```

```python
data_d = match(300)

data_d
```

```python
temp = list(range(200, 2300+1, 100))
n_diff = []
n_melt = []
n_ins = []

for t in temp:
    df = match(t)
    n_diff.append(len(df[df[str(t)]>0]))
    n_melt.append(df[str(t)].isnull().sum())
    n_ins.append(len(df[df[str(t)]==0]))
    
plt.plot(temp, n_diff, label='Number of conductors')
print(n_diff)
print(temp)
plt.plot(temp, n_melt, label='Number of melted materials')
plt.plot(temp, n_ins, label='Number of insulators')
plt.xlabel('Temperature [K]')
plt.ylabel('Number of observations')
plt.legend(frameon=False)
plt.show()
```
